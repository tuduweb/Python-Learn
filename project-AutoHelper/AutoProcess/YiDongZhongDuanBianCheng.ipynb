```python
homework = pd.read_excel('N:/研究生/助教工作/移动终端编程/实践成绩.xlsx',header=0,usecols={"真实姓名","学号","个人总成绩"})
homework
homework_grade = {}
for index,row in homework.iterrows():
    if homework.at[index, '个人总成绩'] > homework_grade.get(homework.at[index, '学号'],0):
        homework_grade[homework.at[index, '学号']] = homework.at[index, '个人总成绩'] 

print(homework_grade)
```

```python
for index,row in sheet.iterrows():
    grade = homework_grade.get(sheet.at[index, '学号'],0)
    if grade >= 4000:
        grade = 4200
    sheet.at[index, '实训作业'] = grade/42

sheet
```

```python
#平时成绩30分中的5分，课堂参与度
for index,row in sheet.iterrows():
    #participation = homework_grade.get(sheet.at[index, '课堂参与度'],"").split(",")
    participation = [int(d) for d in str(sheet.at[index, '课堂参与度']).split(",")]
    grade = 0
    if len(participation) >= 3 and sum(participation) >= 100:
        grade = 100
    elif len(participation) >= 3:
        grade = 90
    elif len(participation) >= 2 or sum(participation) >= 60:
        grade = 80
    elif len(participation) == 1 and participation[0] > 10:
        grade = 70
    elif len(participation) == 1 and participation[0] > 0:
        grade = 60
    sheet.at[index, '课堂成绩'] = grade

sheet
```

```python
presentation = pd.read_excel('N:/研究生/助教工作/移动终端编程/答辩成绩.xls',header=0,usecols={"学号1","学号2","学号3","基础","UI","发挥","最终提交的文档","小组序号","分数4备注"})
presentation = presentation.fillna(0)
presentation = presentation.astype({'学号2': 'int64', '学号3': 'int64'})
presentation
```

```python
presentation_grade = {}
for index,row in presentation.iterrows():
    if presentation.at[index, '学号1'] > 0:
        presentation_grade[presentation.at[index, '学号1']] = [ presentation.at[index, '基础'], presentation.at[index, 'UI'], presentation.at[index, '发挥'], presentation.at[index, '最终提交的文档'], presentation.at[index, '小组序号'], presentation.at[index, '分数4备注']]
    
    if presentation.at[index, '学号2'] > 0:
        presentation_grade[presentation.at[index, '学号2']] = [ presentation.at[index, '基础'], presentation.at[index, 'UI'], presentation.at[index, '发挥'], presentation.at[index, '最终提交的文档'], presentation.at[index, '小组序号'], presentation.at[index, '分数4备注']]
    
    if presentation.at[index, '学号3'] > 0:
        presentation_grade[presentation.at[index, '学号3']] = [ presentation.at[index, '基础'], presentation.at[index, 'UI'], presentation.at[index, '发挥'], presentation.at[index, '最终提交的文档'], presentation.at[index, '小组序号'], presentation.at[index, '分数4备注']]

print(len(presentation_grade))
print(presentation_grade)
```

```python
sheet = sheet.astype({'小组文档备注': 'str', "小组序号":'int64'})

for index,row in sheet.iterrows():
    grades = presentation_grade.get(sheet.at[index, '学号'])
    if grades is not None:
        sheet.at[index, "基础"] = grades[0]
        sheet.at[index, 'UI'] = grades[1]
        sheet.at[index, '发挥'] = grades[2]
        sheet.at[index, '提交的文件'] = grades[3]
        sheet.at[index, '小组序号'] = grades[4]
        sheet.at[index, '小组文档备注'] = grades[5]
    else:
        sheet.at[index, '小组序号'] = -1
        sheet.at[index, '小组文档备注'] = "未参加答辩"

sheet = sheet.fillna(0)
sheet = sheet.astype({'学号': 'int64'})
        
sheet
```
